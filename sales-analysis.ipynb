```python
#Now create a copy so we don't damage raw data:
clean_df1 = df1.copy()
clean_df2 = df2.copy()
```

```python
clean_df1.isnull().sum()
```

```python
clean_df2.isnull().sum()
```

```python
clean_df1 = clean_df1.drop_duplicates()

clean_df2 = clean_df2.drop_duplicates()
```

```python
clean_df1.info()
```

```python
clean_df2.info()
```

```python
#Sometimes CSVs have hidden spaces.
clean_df1.columns = clean_df1.columns.str.strip()
clean_df2.columns = clean_df2.columns.str.strip()
```

```python
#Clean text columns make it in for loop

text_columns = [
    "fuel",
    "vehicleClass",
    "category",
    "OfficeCd",
    "secondVehicle",
    "tempRegistrationNumber"
]


for col in text_columns:
    clean_df1[col] = clean_df1[col].str.strip().str.upper()
    clean_df2[col] = clean_df2[col].str.strip().str.upper()
```

```python
date_columns = [
    "makeYear",
    "insuranceValidity",
    "fromdate",
    "todate"
]


for col in date_columns:

    clean_df1[col] = pd.to_datetime(
        clean_df1[col],
        format="%d/%m/%y",
        errors="coerce"
    )


    clean_df2[col] = pd.to_datetime(
        clean_df2[col],
        format="%d/%m/%y",
        errors="coerce"
    )
```

```python
clean_df1.info()
```

```python
clean_df1["fuel"].value_counts()
```

```python
#FILL WITH MOST Common value
clean_df1["fuel"] = clean_df1["fuel"].replace(
    {
        "PETROL CNG": "CNG PETROL",
        "+AC0-1": "UNKNOWN"
    }
)


clean_df2["fuel"] = clean_df2["fuel"].replace(
    {
        "PETROL CNG": "CNG PETROL",
        "+AC0-1": "UNKNOWN"
    }
)
```
